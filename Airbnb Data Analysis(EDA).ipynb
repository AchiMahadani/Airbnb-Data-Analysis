```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Airbnb_Open_Data.csv")
```

```python
df.info()
```

```python
df
```

```python
df.columns
```

```python
print(df.isnull().sum())
```

```python
df['last review'] = pd.to_datetime(df['last review'],errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month': 0, 'last review': df['last review'].min()} ,  inplace = True)
```

```python
df.head()
```

```python
df.dropna(subset = ['NAME' , 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=['house_rules' , 'license'],errors = 'ignore')
```

```python
df.head()
```

```python
## Remove dollar signs and convert to float

df['price'] = df['price'].replace('[\$,]', '',regex=True).astype(float)

df['service fee'] = df['service fee'].replace('[\$]','',regex=True).astype(float)
```

```python
df.head()
```

# Removing Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

## What is the distribution of listing prices?

```python
plt.figure(figsize =(10,6))
sns.histplot(df['price'],bins=50,kde=True,color='Red')
plt.title("Distribution of listing price")
plt.xlabel("price $")
plt.ylabel("Frequency")
plt.show()


```

**The histogram shows a fairly distribution of listing prices across different price ranges,indicating no**
**particular concentration of listings in any specific price range.**
**The KDE line helps visualise this even spread more clearly, confirming that the dataset contains listings with a wide 
variety of prices.**

## How are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x = "room type", data = df, color="Purple")
plt.title("Room Type Distribution")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()
```

**Conclusion: Mostly Entire home preferred by customer and then private room.**

## How are Listings distributed across different neighborhoods?

```python
plt.figure(figsize=(12,8))
sns.countplot(y='neighbourhood group' , data=df ,color ="lightgreen",order =df['neighbourhood group'].value_counts().index)
plt.title("Number of Lsitings by Neighbourhood Group")
plt.xlabel("Count")
plt.ylabel('Neighborhood Group')
plt.show()
```

## What is the relationship between price and room type?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type' , y='price', hue= 'room type',data =df , palette ='Set1')
plt.title('Price vs Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price($)')
plt.legend(title='Room Type')
plt.show()
```

## How has the number of reviews change over with time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel("Date")
plt.ylabel("Number of Reviews")
plt.show()
```
